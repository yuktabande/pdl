```python
# 1. Import required libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
import torchvision.transforms as transforms
from PIL import Image
```

```python
# 2. Load and display the input image
image_path = '/content/image.jpg'
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5, 5))
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
# 3. Enhance Image Quality - Brightness & Contrast
# Brightness & Contrast adjustment
brightness = 50
contrast = 30
enhanced_img = cv2.convertScaleAbs(img_rgb, alpha=1 + contrast/100.0, beta=brightness)

plt.figure(figsize=(5, 5))
plt.title("Brightness & Contrast Enhanced")
plt.imshow(enhanced_img)
plt.axis('off')
plt.show()
```

```python
# 4. Data Augmentation

# Convert to PIL for torchvision transforms
img_pil = Image.fromarray(img_rgb)

augmentations = transforms.Compose([
    transforms.RandomHorizontalFlip(p=1.0),
    transforms.RandomRotation(30),
    transforms.RandomResizedCrop(size=224, scale=(0.8, 1.0))
])

aug_img = augmentations(img_pil)

plt.figure(figsize=(5, 5))
plt.title("Augmented Image")
plt.imshow(aug_img)
plt.axis('off')
plt.show()
```
