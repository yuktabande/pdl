Step 1: Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Step 2: Load and Prepare the Dataset

```python
data = pd.read_csv("/kaggle/input/students-score-dataset-linear-regression/student_scores.csv")  # Replace with your file name
X = data["Hours"].values.reshape(-1, 1)   # Convert to 2D array
y = data["Scores"].values.reshape(-1, 1)
```

Step 3: Normalize the Data

```python
X = X / X.max()
y = y / y.max()
```

Step 4: Initialize Weights and Biases

```python
np.random.seed(42)
w = np.random.rand(1)  # One weight (since we have one feature)
b = np.random.rand(1)  # One bias
```

Step 5: Define Hyperparameters

```python
learning_rate = 0.1
epochs = 1000
```

 Step 6: Activation Function (Linear Model)

```python
def predict(X):
    return X * w + b
```

Step 7: Loss Function (MSE)

```python
def mse(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)
```

Step 8: Training Loop (Gradient Descent)

```python
losses = []

for epoch in range(epochs):
    y_pred = predict(X)
    loss = mse(y_pred, y)
    losses.append(loss)

    # Gradients
    dw = np.mean(2 * X * (y_pred - y))
    db = np.mean(2 * (y_pred - y))

    # Update parameters
    w -= learning_rate * dw
    b -= learning_rate * db

    # Optional: print every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}")
```

Step 9: Plot Loss vs Epoch

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.title("Training Loss over Epochs")
plt.show()
```

Step 10: Test Model on New Input

```python
test_hours = np.array([[5]]) / X.max()  # Normalize same as training
predicted_score = predict(test_hours) * y.max()  # Un-normalize output

print(f"Predicted score for 5 hours: {predicted_score[0][0]:.2f}")
```
