```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
image = cv2.imread('/content/img.webp')
```

```python
cv2_imshow(image)
```

Grayscale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

Histogram Equalization (for better contrast)

```python
ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
ycrcb[:, :, 0] = cv2.equalizeHist(ycrcb[:, :, 0])
equalized = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)
cv2_imshow(equalized)
```

Contrast stretching (normalize pixel values)

```python
normalized = cv2.normalize(image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
cv2_imshow(normalized)
```

Image Smoothing(Blurring)

```python
blurred = cv2.GaussianBlur(image, (7, 7), 0)
cv2_imshow(blurred)
```

Image Sharpening

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
sharpened = cv2.filter2D(image, -1, kernel)
cv2_imshow(sharpened)
```

horizontal flip

```python
h_flip = cv2.flip(image, 1)
cv2_imshow(h_flip)
```

vertical flip

```python
v_flip = cv2.flip(image, 0)
cv2_imshow(v_flip)
```

rotation

```python
(h,w) = image.shape[:2]
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
cv2_imshow(rotated)
```

translation

```python
M = np.float32([[1, 0, 50], [0, 1, 30]])
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
cv2_imshow(shifted)
```

Scaling

```python
resized = cv2.resize(image, None, fx=1.5, fy=1.5)
cv2_imshow(resized)
```
